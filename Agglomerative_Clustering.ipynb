```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= pd.read_csv('attrition2.csv')
data
```

```python
modified_data = data.drop(['Emp ID','Work_accident','promotion','dept','salary'],axis=1)
correlation = modified_data.corr()
correlation
```

```python
modified_data
```

```python
X = modified_data.iloc[:,[1,3]].values
X
```

```python
from sklearn.cluster import AgglomerativeClustering
aggl = AgglomerativeClustering(n_clusters = 2)
```

```python
assign = aggl.fit_predict(X)
plt.scatter(X[:,0],X[:,1], c=assign)
plt.xlabel('last_evaluation')
plt.ylabel('average_hours')

plt.show()
```
